```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
points = np.arange(-5, 5, 0.01)
```

```python
dx, dy = np.meshgrid(points, points)
```

```python
dx
```

```python
dy
```

```python
x = np.sin(dx)
y = np.sin(dy)

z = (np.sin(dx) + np.sin(dy))
x
```

```python
plt.imshow(z)
```

```python
plt.imshow(x)
plt.colorbar()
plt.title('Plot for sin(x)')
```

```python
plt.imshow(y)
plt.colorbar()
plt.title('Plot for sin (y)')
```

```python
# numpy where

A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
condition = np.array([True, True, False, False])
```

```python
answer = [(A_val if cond else B_val) 
          for A_val, B_val, cond in zip(A,B,condition)]
```

```python
answer
```

```python
answer2 = np.where(condition, A, B)
```

```python
answer2
```
