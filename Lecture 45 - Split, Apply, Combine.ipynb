```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe_wine = pd.read_csv('winequality-red.csv', sep=';')
dframe_wine.head()
```

```python
dframe_wine_q3 = dframe_wine[dframe_wine.quality == 3]
```

```python
dframe_wine_q3_sorted = dframe_wine_q3.sort('alcohol', ascending=False)
dframe_wine_q3_sorted.head()
```

```python
dframe_wine_q3_sorted['alc_content_rank'] = np.arange(10) + 1
dframe_wine_q3_sorted
```

```python
def ranker(df):
    print('Lengh of dframe {}: {}'.format(df.__name__, len(df)))
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort('alcohol', ascending=False, inplace=True)
dframe_wine.head()
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
dframe_wine.head()
```

```python
num_of_qual = dframe_wine['quality'].value_counts()
```
